```python
%%writefile arithmetic1.c

#include <stdio.h>

int main () {
    int x,y,result;
    printf("두개의 정수를 입력 하시오: ");
    scanf("%d %d",&x,&y);

    result = x + y;
    printf("%d + %d = %d\n",x,y,result);

    result = x - y;
    printf("%d - %d = %d\n",x,y,result);
    result = x * y;
    printf("%d * %d = %d\n",x,y,result);
    result = x / y;
    printf("%d / %d = %d\n",x,y,result);
    result = x % y;
    printf("%d %% %d = %d\n",x,y,result);
    
    return 0;
}
```

```python
%%shell
gcc arithmetic1.c -o output
./output
```

```python
%%writefile modulo.c

#include <stdio.h>
#define SEC_PER_MINUTE 60

int main () {
    int input, minute, second;

    printf("초단위의 시간을 입력하시오: (32억초 이하)");
    scanf("%d",&input);

    minute = input / SEC_PER_MINUTE;
    second = input % SEC_PER_MINUTE;

    printf("%d초는 %d분 %d초 입니다.\n",input, minute, second);

    return 0;
}
```

```python
%%shell
gcc modulo.c -o output
./output
```

```python
%%writefile plus.c

#include <stdio.h>

int main () {
    int x,y;
    x = 1;
    y = ++x;
    printf("x = %d y =  %d \n",x,y);

    y = x++;
    printf("x = %d y =  %d \n",x,y);

    return 0;
}
```

```python
%%shell
gcc plus.c -o output
./output
```

```python
%%writefile bit.c

#include <stdio.h>

int main () {
    int x = 9;
    int y = 10;
    
    printf("%08X & %08X = %08X \n",x, y, x & y); // 8자리의 16진수로 표시하라
    printf("%08X | %08X = %08X \n",x, y, x | y); // 8자리의 16진수로 표시하라
    printf("%08X ^ %08X = %08X \n",x, y, x ^ y); // 8자리의 16진수로 표시하라
    printf("~%08X = %08X \n",x, ~x);  // 8자리의 16진수로 표시하라
    
    return 0;
}
```

```python
%%shell
gcc bit.c -o output
./output
```

```python
%%writefile order.c

#include <stdio.h>

int main () {
    int a = 10;
    int b = 20; 
    int c = 30;
    int d = 3;
    int result;

    result = a + b * c / d;  //10 + 20*30/3 = 10 + 200
    printf("연산값 : %d \n",result); // 30 * 30 / 3 
    result = (a + b) * c / d;
    printf("연산값 : %d \n",result);
    result = a = b = 1;;
    printf("연산값 : %d \n",result);

    return 0;
}
```

```python
%%shell
gcc order.c -o output
./output
```

```python
%%writefile conv.c

#include <stdio.h>

int main () {
    int i;
    double f;

    f = 5/4;
    printf("(5/4) = %lf \n",f);
    f = (double)5 / 4;
    printf("(double)5 / 4 = %lf \n",f);

    i = 1.3 + 1.8;
    printf("1.3 + 1.8 = %d \n",i); 
    i = (int)1.3 + (int)1.8; 
    printf("(int)1.3 + (int)1.8 = %d \n",i);

    return 0;
}
```

```python
%%shell
gcc conv.c -o output
./output
```
