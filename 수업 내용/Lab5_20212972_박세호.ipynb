```python
%%writefile temp.c

#include <stdio.h>

int main () {
     int temperature;
     printf("온도를 입력하시오: ");
     scanf("%d", &temperature);

    if (temperature > 0)
        printf("영상의 날씨입니다. \n");
    else
        printf("영하의 날씨입니다. \n");

    printf("현재 온도는 %d도 입니다.\n",temperature);
    return 0;
}
```

```python
%%shell
gcc temp.c -o output
./output
```

```python
%%writefile block.c

#include <stdio.h>

int main () {
     int score;

     printf("성적을 입력하시오: ");
     scanf("%d", &score);

    if (score >= 60) {
        printf("합격입니다. \n");
        printf("장학금도 받을 수 있습니다. \n");
    }
    
    else {
        printf("불합격입니다. \n");
        printf("다시 도전하세요. \n");
    }

    return 0;
}
```

```python
%%writefile big.c

#include <stdio.h>

int main () {

    int x, y;

    printf("첫번째 수: ");
    scanf("%d",&x);
    printf("두번째 수: ");
    scanf("%d",&y);

    printf("큰 수: %d\n",(x>y)? x:y);
    printf("작은 수: %d\n",(x<y)? x:y);

    return 0;
}
```

```python
%%shell
gcc big.c -o output
./output
```

```python
%%writefile grade.c

#include <stdio.h>

int main () {
    int score;
    char grade;

    printf("성적을 입력하십시오: ");
    scanf("%d",&score);

    if (score >= 90)
        grade = 'A';
    else if (score >= 80)
        grade = 'B';
    else if (score >= 70)
        grade = 'C';
    else if (score >= 60)
        grade = 'D';
    else
        grade = 'F'; // 위에 다 거짓이면 이것이 실행됨

    printf("학점: %c \n",grade);

    return 0;
}
```

```python
%%shell
gcc grade.c -o output
./output
```

```python
%%writefile operator.c

#include <stdio.h>

int main () {
    char op;
    int x,y;

    printf("수식을 입력하십시오: ");
    scanf("%d %c %d",&x, &op, &y);

    if (op == '+') 
        printf("%d\n",x+y);
    else if (op == '-') 
        printf("%d\n",x-y);
    else if (op == '*') 
        printf("%d\n",x*y);
    else if (op == '/') 
        printf("%d\n",x/y);
    else 
        printf("지원되지않는 연산자입니다.\n");

    return 0;
}
```

```python
%%shell
gcc operator.c -o output
./output
```

```python
%%writefile month.c

#include <stdio.h>

int main () {
    int month, days;
    printf("일수를 알고 싶은 달을 입력하시오: ");
    scanf("%d",&month);

    switch (month) 
    {
        case 2:
            days = 28;
            break;
        case 4:
        case 6:
        case 9:
        case 11:
            days = 30;
            break;

        default:
            days = 31;
            break;
    }

    printf("%d월의 일수는 %d입니다.\n", month, days);
    return 0;
}
```

```python
%%shell
gcc month.c -o output
./output
```
