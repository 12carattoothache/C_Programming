```python
%%writefile sum3.c

#include <stdio.h>

int main(void)
{
  int x;
  int y;
  int sum;

  printf("첫 번째 숫자를 입력하시오 : ");
  scanf("%d", &x);  // 하나의 정수를 받아서 x에 저장

  printf("두 번째 숫자를 입력하시오 : ");
  scanf("%d", &y);  // 하나의 정수를 받아서 y에 저장

  sum = x + y;
  printf("두 수의 합 = %d\n", sum);

  return 0;
}
```

```python
ls
```

```python
%%shell
gcc sum3.c -o outsum3
./outsum3
```
