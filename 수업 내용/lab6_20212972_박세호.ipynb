```python
%%writefile repeat.c

#include <stdio.h>

int main (void) {
    int i = 0;
    while (i < 10) {
        printf("수업 시간에 떠들지 않겠습니다\n");
        printf("i는 %d\n", i);
        i++;
    }
    return 0;
}
```

```python
%%shell
gcc repeat.c -o output
./output
```

```python
%%writefile mile.c

# include <stdio.h>

int main (void) {
  
    int meter;
    int i = 0;

    while (i <= 10) {
        meter = i * 1609;
        printf("%d마일은 %d미터입니다.\n",i,meter);
        i++;
    }
    return 0;
}
```

```python
%%shell
gcc mile.c -o output
./output
```

```python
%%writefile fact.c

# include <stdio.h>

int main (void) {
    int i = 10;
    int factorial = 1;

    while (i>=1) {
        factorial *= i;
        i--;
    }    
    
    printf("%d\n",factorial);

    return 0;
}
```

```python
%%shell
gcc fact.c -o output
./output
```

```python
%%writefile cal.c

# include <stdio.h>

int main (void) {
    int n, i = 1;

    printf("구구단 중에서 출력하고 싶은 단을 입력하시오: ");
    scanf("%d",&n);

    while (i <= 9) {
        printf("%d * %d = %d\n",n,i,n*i);
        i++;
    }
    return 0;
}
```

```python
%%writefile sum.c

# include <stdio.h>

int main (void) {
    int i = 1;
    int sum = 0;

    while (i <= 1000) {
        sum += i;
        i++;
    }
    printf("합은 %d입니다.\n",sum);

    return 0;
}
```

```python
%%writefile dowhile.c

# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>

int main (void) {
    int i;
    do {
        printf("1----파일 열기\n");
        printf("2----파일 저장하기\n");
        printf("3----종료\n");
        printf("하나를 선택하시오 : ");
        scanf("%d",&i);
    } while(i < 1 || i > 3);

    printf("선택된 메뉴 = %d\n",i);

    return 0;
}
```

```python
%%shell
gcc dowhile.c -o output
./output
```

```python
%%writefile sum.c

# include <stdio.h>

int main (void) {

    int i, sum = 0;

    for (i = 1; i <= 50; i++) // i는 1부터 10까지 증가
    {
        sum += i; // sum = sum + i 와 같음
    }

    printf("1부터 50까지의 정수의 합 = %d\n",sum);
    
    return 0;
}
```

```python
%%shell
gcc sum.c -o output
./output
```

```python
%%writefile square.c

// 반복을 이용한 네모 그리기
# include <stdio.h>

int main (void) {
    int i;
    printf("*********\n");

    for(i = 0; i<5; i++) {
        printf("*       *\n");
    }

    printf("*********\n");;
    return 0;
}
```

```python
%%writefile fact.c

# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>

int main (void) {

    int fact = 1;
    int i, n;

    printf("정수를 입력하시오: ");
    scanf("%d", &n);

    for (i=1; i <= n; i++) {
        fact = fact * i;
    }

    printf("%d!은 %d입니다.\n",n,fact);
    return 0;
}
```

```python
%%shell
gcc fact.c -o output
./output
```

```python
%%writefile nest.c

// 중첩 for문을 이용하여 *기호를 사각형 모양으로 출력하는 프로그램
# include <stdio.h>

int main (void) {

    int x, y;

    for (y = 0; y < 5; y++)    // 바깥 반복문
    {
        for (x = 0; x < 10; x++) {  // 안쪽 반복문
            printf("*");
        }
        printf("\n"); // 안쪽 반복문이 종료될 때마다 실행
    }

    return 0;
}
```

```python
%%shell
gcc nest.c -o output
./output
```

```python
%%writefile sum.c

# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>

int main (void) {

    int i;
    double number, sum = 0.0;

    for (i = 1; i <= 10; i++) {
        printf("%d번째 실수를 입력하시오: ",i);
        scanf("%lf", &number);

        if (number < 0.0) {
            break;
        }
        sum += number;
    }

    printf("합계: %lf",sum);

    return 0;
}
```

```python
%%shell
gcc sum.c -o output
./output
```

```python
%%writefile continue.c

# include <stdio.h>

int main (void) {

    int i;

    for (i = 0; i < 10; i++) {
        if (i % 2 == 1) {
            continue;
        }
        printf("정수: %d \n", i);
    }
    return 0;
}
```

```python
%%shell
gcc continue.c -o output
./output
```
