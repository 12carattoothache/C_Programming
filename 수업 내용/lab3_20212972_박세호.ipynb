```python
%%writefile overflow.c

#include <stdio.h>

int main(void)
{
  short s_money = 32767; //max number of short type

  s_money = s_money + 1;
  printf("s_money = %d\n", s_money);

  return 0;
}
```

```python
%%shell

gcc overflow.c -o overflow_out
./overflow_out
```

```python
%%writefile ascii1.c

#include <stdio.h>

int main(void) {
  char c;

  c = 'A';
  printf("A의 아스키코드 = %d\n", c);

  printf("문자를 입력하시오: ");
  c = getchar();

  printf("%c의 \"아스키\" 코드 = %d", c, c);

  return 0;
}
```

```python
%%shell

gcc ascii1.c -o output
./output
```

```python
%%writefile idpw.c

#include <stdio.h>

int main(void) {
  int id, pass;

  printf("아이디와 패스워드를 4개의 숫자로 입력하세요: \n");
  printf("id : _______   \b\b\b\b");
  scanf("%d", &id);

  printf("pass : _______   \b\b\b\b");
  scanf("%d", &pass);

  printf("\n입력된 아이디는 \"%d\"이고 패스워드는 \"%d\"입니다.\n", id, pass);

  return 0;
} 
```

```python
%%shell

gcc idpw.c -o output
./output
```
