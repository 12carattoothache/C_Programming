2-1 4칙연산(1)

```python
%%writefile lab2-1.c

#include <stdio.h>

int main(void)
{
  int x;
  int y;
  int sum, sub, mul, div;

  printf("첫 번째 숫자를 입력하시오 : ");
  scanf("%d", &x);  // 하나의 정수를 받아서 x에 저장

  printf("두 번째 숫자를 입력하시오 : ");
  scanf("%d", &y);  // 하나의 정수를 받아서 y에 저장

  sum = x + y;
  sub = x - y;
  mul = x * y;
  div = x / y;

  printf("두 수의 합 = %d\n", sum);
  printf("두 수의 차 = %d\n", sub);
  printf("두 수의 곱 = %d\n", mul);
  printf("두 수의 몫 = %d\n", div);

  return 0;
}
```

```python
%%shell

gcc lab2-1.c -o output1
./output1
```

2 - 2 4칙연산(2)

```python
%%writefile lab2-2.c

#include <stdio.h>

int main(void)
{
    float x;
    float y;
    float evr;
 
    printf("첫 번째 상품의 가격을 입력하십시오: ");
    scanf("%f",&x);
    printf("두 번째 상품의 가격을 입력하십시오: ");
    scanf("%f",&y);

    evr = (x+y)/2 ;
    printf("두 상품의 평균 가격: %f\n",evr);
 
    return(0);
}
```

2 - 3 여행비용 계산

```python
%%writefile lab2-3.c

#include <stdio.h>

int main(void)
{
    int a,b,c,d,sum;
 
    printf("여행은 몇박인가요?");
    scanf("%d",&a);
    printf("항공권 가격: ");
    scanf("%d",&b);
    printf("호텔 1박 가격: ");
    scanf("%d",&c);
    printf("하루에 필요한 용돈 : ");
    scanf("%d",&d);
    printf("======================\n");

    sum = a*c + b + a*d;
    printf("총 여행 비용: %d\n",sum);
    
    printf("======================\n")

    return(0);
}
```

```python
%%shell

gcc lab2-3.c -o output3
./output3
```
