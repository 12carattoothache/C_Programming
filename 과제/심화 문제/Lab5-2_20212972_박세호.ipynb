```python
%%writefile 1.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main (void) {
    printf("동전 던지기 게임을 시작합니다.\n");
    srand(time(NULL)); // 현재 시간을 난수의 씨앗값으로 한다.

    int coin = rand() % 2;

    if (coin == 0){
        printf("앞면입니다");
    }
    else {
        printf("뒷면입니다");
    }

    return 0;
}
```

```python
%%writefile 3.c

#define _CRT_SECURE_NO_WARNINGS
#include <stdio.h>

int main (void) {
    int x,y,z,max;

    printf("3개의 정수를 입력하시오: ");
    scanf("%d %d %d",&x,&y,&z);

    max = x;

    if(max < y){
        max = y;
    }

    if(max < z){
        max = z;
    }

    printf("가장 큰 정수는 %d이다.\n",max);

    return 0;
}
```

```python
%%writefile 3-1.c

#include <stdio.h>

int main (void) {
    int x,y,z,min;

    printf("3개의 정수를 입력하시오: ");
    scanf("%d %d %d",&x,&y,&z);

    min = (x > y) ? y : x;
    min = (min > z) ? z : min;

    printf("가장 작은 정수는 %d이다.\n",min);

    return 0;
}
```

```python
%%writefile 6.c

#define _CRT_SECURE_NO_WARNINGS
#include <stdio.h>
#include <math.h>

int main (void) {
  
    int a, b, c;
    double d, x1, x2;
    
    printf("계수 a,b,c 를 입력하시오: ");
    
    scanf("%d %d %d", &a, &b, &c);
    d = b * b - 4. * a * c;
    
    if (d == 0) {
        x1 = x2 = (-b) / (2. * a);
        printf("실근(중근): %f\n", x1);
    }

    else if (d > 0) {
        x1 = (-b + sqrt(d)) / (2. * a);
        x2 = (-b - sqrt(d)) / (2. * a);
        printf("첫 번째 실근: %f\n", x1);
        printf("두 번째 실근: %f\n", x2);
    }

    else {
        printf("허수근입니다.\n");
    }

    return 0;
}
```

```python
%%shell
gcc 6.c -o output -lm
./output
```

#Mini Project: 가위 바위 보 게임

```python
%%writefile 7.c

#include<stdio.h>
#include<stdlib.h>
#include<time.h>

int main (void) {
        int com, human;
        printf("가위 바위 보 게임에 오신 것을 환영합니다.\n");
        printf("하나를 선택하세요(가위: 0, 바위: 1, 보: 2): ");
        scanf("%d", &human);

        srand(time(NULL));
        com = rand() % 3;

        printf("사용자: %d \n컴퓨터: %d \n",human,com);

        if (human == 0 && com == 2) {
            printf("컴퓨터 승리\n");
        }

        else if (human == 2 && com == 0) {
            printf("사용자 승리\n");
        }

        else if (human > com) {
            printf("사용자 승리\n");
        }

        else if (human < com) {
            printf("컴퓨터 승리\n");
        }

        else if (human == com) {
            printf("비겼습니다\n");
        }

        else {
            printf("에러!\n");
        }
    return 0;
    }
```

```python
%%shell
gcc 7.c -o output -lm
./output
```
