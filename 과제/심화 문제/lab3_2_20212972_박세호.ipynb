Lab 3-1

```python
%%writefile lab3_1.c
#include<stdio.h>

#define exchange_rate 1120

int main(void){
    int dollar,won;
    printf("달러화 금액을 입력하시오 : ");
    scanf("%d",&dollar);
    won = dollar * exchange_rate;
    printf("%d달러는 %d원입니다.",dollar,won);

    return 0;
}
```

```python
%%shell
gcc lab3_1.c -o output
./output
```

Lab 3-5

```python
%%writefile lab3_5.c
#include <stdio.h>

int main(void){
    double lightspeed = 300000;
    double distance = 149600000;
    double time;

    time = distance / lightspeed;
    time = time/60;

    printf("빛의 속도는 %lfkm/s \n",lightspeed);
    printf("태양과 지구 사이의 거리는 %lfkm \n",distance);
    printf("도달시간은 %lf분 \n",time);
    
    return 0;
}
```

```python
%%shell
gcc lab3_5.c -o output
./output
```

```python
%%writefile lab3_6.c
#include<stdio.h>

int main(void){
    float fahrenheit,celsius;

    printf("화씨온도: ");
    scanf("%f",&fahrenheit);

    celsius = (fahrenheit - 32)*5/9;
    printf("섭씨온도: %f",celsius);
    return 0;
}
```

Lab 3-7

```python
%%writefile lab3_7.c
#include<stdio.h>
#define pi 3.141592

int main(void){
    float radius,area;
    printf("원의 반지름을 입력하시오: ");
    scanf("%f",&radius);
    area = pi * radius * radius;
    printf("원의 넓이 : %f",area);
    
    return 0;
}
```

```python
%%shell
gcc lab3_7.c -o output
./output
```

Lab 3-8 (mini project)

```python
%%writefile lab3_8.c
#include<stdio.h>

int main(void){
    short age,id;
    long products;
    double stars;
    char character;

    printf("나이를 입력하시오 : ");
    scanf("%hi",&age);
    printf("나이: %hi \n",age);

    printf("직원 아이디를 입력하시오 : ");
    scanf("%hi",&id);
    printf("아이디: %hi \n",id);

    printf("생산량을 입력하시오 : ");
    scanf("%ld",&products);
    printf("생산량: %ld \n",products);

    printf("별의 개수를 입력하시오 : ");
    scanf("%lf",&stars);
    printf("별의 개수: %e \n",stars);

    printf("글자를 입력하시오 : ");
    getchar();
    character = getchar();
    printf("글자: %c \n",character);

    return 0;
}
```

```python
%%shell
gcc lab3_8.c -o output
./output
```
