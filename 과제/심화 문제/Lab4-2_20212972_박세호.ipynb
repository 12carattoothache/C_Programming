# Lab 4-1

```python
%%writefile 1.c

#include <stdio.h>

int main () {
    int days, weeks, years;

    printf("총 일수를 입력 하시오: ");
    scanf("%d", &days);

    years = days / 365;
    weeks = (days % 365) / 7;
    days = days - (years * 365) - (weeks * 7);
    printf("%d년", years);
    printf("%d주", weeks);
    printf("%d일\n", days);

    return 0;
}
```

```python
%%shell
gcc 1.c -o output
./output
```

# Lab 4-2

```python
%%writefile 2.c

#include <stdio.h>

int main () {
    int input, price, change, coin_hundred, coin_ten;

    printf("투입한 돈 : ");
    scanf("%d",&input);
    printf("물건값 : ");
    scanf("%d",&price);

    change = input - price;
    printf("거스름 돈: %d \n", change);

    coin_hundred = change / 100;
    change %= 100;
    coin_ten = change / 10;

    printf("100원의 개수: %d \n10원의 개수: %d",coin_hundred, coin_ten);

    return 0;
}
```

```python
%%shell
gcc 2.c -o output
./output
```

# Mini Project: 복리계산

```python
%%writefile 5.c

#include <stdio.h>
#include <math.h> 

int main () {
    double a1, an, rate;
    int year;

    printf("초기저금액: ");
    scanf("%lf", &a1);
    printf("이율: ");
    scanf("%lf", &rate);
    printf("저축기간: ");
    scanf("%d", &year);
    
    
    an = a1 * pow(1+rate , year); 

    printf("%d년 후의 원리금: %lf \n",year,an);
    
    return 0;
}
```

```python
%%shell
gcc 5.c -o output -lm
./output
```
