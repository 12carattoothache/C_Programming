# Lab 6-2

```python
%%writefile  6-2.c

# include <stdio.h>
# include <stdlib.h>

int main (void) {
    int h = 0;
    int m = 0;
    int s = 0;
  
    while(1) {
        printf("%02d: %02d: %02d", h, m, s);

        s++;

        if (s == 60) {
            m ++;
            s = 0;
        }

        if (m == 60) {
            h ++;
            m = 0;
        }
        if (h == 24) {
          h = m = s = 0;
        }
    
        sleep(1000); // 1초 동안 프로그램을 재운다.
    }
    return 0;
}
```

```python
%%shell
gcc 6-2.c -o output -lm
./output
```

# Lab 6-3

```python
%%writefile 6-3.c

# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>
# include <stdlib.h>
# include <time.h>

int main (void) { 
    srand(time(NULL));
    
    int answer = rand() % 100; //정답
    int guess;
    int tries = 0;
    // 반복구조
    do {
        printf("정답을 추측하여 보시오(0~100): ");
        scanf("%d", &guess);
        tries++;

        if (guess > answer) {
        printf("제시한 정수가 높습니다.\n");
        }
        else if (guess < answer) {
        printf("제시한 정수가 낮습니다.\n"); 
        }
    } while (guess != answer);
    
    printf("축하합니다. 시도횟수 = %d \n", tries);

    return 0;
}
```

```python
%%shell
gcc 6-3.c -o output -lm
./output
```

# Lab 6-4

```python
%%writefile 6-4.c
# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>

int main (void) {
    long long num; // 64비트형 정수형
    int count = 0;
    
    printf("정수를 입력하시오: ");
    scanf("%lld", &num); // long long 타입은 %lld를 사용해야 한다.
    
    do {
        count++;
        num /= 10; // 정수 나눗셈 
    } while (num != 0); // 일의 자리 수를 10으로 나누면 0

    printf("총 자리수: %d", count);
    return 0;
}
```

```python
%%shell
gcc 6-4.c -o output
./output
```

# Lab 6-8

```python
%%writefile 6-8.c

# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>
# include <stdlib.h>
# include <time.h>

int main (void) {

    srand(time(NULL));

    int i;
    int ans;

    printf("산수 문제를 자동으로 출제합니다. \n");

    while (1) {
        int x = rand() % 100; //0~99 사이의 난수
        int y = rand() % 100; //0~99 사이의 난수
        
        printf("%d + %d = ", x, y);
        scanf("%d", &ans);

        if (x + y == ans) {
            printf("맞았습니다.\n");
            break;
        }
        else {
            printf("틀렸습니다.\n");
        }
    }

    return 0;
}
```

# Lab 6-9

```python
%%writefile 6-9.c

# define _CRT_SECURE_NO_WARNINGS
# include <stdio.h>

int main (void) {

    int n, is_prime;

    printf("어디까지 찾을까요? : ");
    scanf("%d", &n);
    
    for (int i=2; i<= n; i++) { // i를 n까지
        is_prime = 1;

        for (int k=2; k<i; k++) { // k를 i 전까지
            if (i % k == 0) { // i가 약수가 존재하면
                is_prime = 0; 
                break;
            }
        }

        if (is_prime == 1) { // i가 약수가 존재하지 않으면
            printf("%d, ", i);
        }
    }
    printf("\b\b");
    return 0;
}
```

```python
%%shell
gcc 6-9.c -o output
./output
```

# Lab 6-10

```python
%%writefile 6-10.c

# define _CRT_SECURE_NO_WARNINGS
#include <stdio.h>
#include <math.h>
#include <stdlib.h>
#include <time.h>


int main (void) {

    srand(time(NULL));
    int niter = 0;
    int inside_dots = 0;
    double x, y, distance, pi;

    printf("반복횟수: ");
    scanf("%d", &niter);

    inside_dots = 0;

    for (int i = 0; i < niter; i++) {
        x = (double)rand() / RAND_MAX; // 1이하의 난수 x
        y = (double)rand() / RAND_MAX; // 1이하의 난수 y
        distance = x * x + y * y;
        if (distance <= 1) { 
            inside_dots++;
        }
    }

    pi = 4.0 * (double)inside_dots / niter;
    printf("파이 = %lf\n", pi);
    return(0);
}
```

```python
%%shell
gcc 6-10.c -o output -lm
./output
```

# Mini Project 6-11 : NIM게임


```python
%%writefile 6-11.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main (void) {
    int i1, i2, remain = 12;

    do {
        printf("현재 스틱의 개수: %d \n",remain);

        // 사용자 턴
        printf("몇개의 스틱을 가져가시겠습니까?(0 ~ 3개까지 가능): ");
        scanf("%d",&i1);

        if (0 <= i1 && i1 <= 3) { 
            if (i1 > remain) {
              printf("현재 스틱의 개수보다 많이 입력할 수 없습니다.\n\n");
              continue;
            }
            
            else {
              remain -= i1;

              if (remain == 0) {
                printf("사용자 패배!\n");
                break;
              }
            }

        // 컴퓨터 턴
            srand(time(NULL));
            i2 = rand()%4; 
            if (i2 > remain) {
                i2 = remain;
            }
            printf("컴퓨터는 %d개의 스틱을 가져갔습니다.\n\n",i2);
            remain -= i2;

            if (remain == 0) { 
                printf("사용자 승리!\n");
                break;
            }
        }

        else {
            printf("범위 내의 수를 입력하세요.\n\n");
        }
    } while (remain > 0);

    return(0);
}
```

```python
%%shell
gcc 6-11.c -o output -lm
./output
```
