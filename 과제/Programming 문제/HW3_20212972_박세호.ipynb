```python
%%writefile HW3-3.c

# include <stdio.h>

# define PI 3.141592

int main (void) {
    double radius, A, V;
    printf("반지름을 입력하시오: ");
    scanf("%lf", &radius);
    A = 4.0 * PI * (radius * radius);
    V = (4.0/3.0) * PI * (radius * radius * radius);
    printf("구의 표면적: %lf\n구의 부피: %lf", A, V);
    return 0;
}
```

```python
%%shell
gcc HW3-3.c -o output
./output
```

```python
%%writefile HW3-8.c

# include <stdio.h>

#define secondtohour 1/3600
#define minutetohour 1/60
#define mtokm 1/1000

int main (void) {
    float distance,hour,minute,second,v;
    printf("거리를 미터로 입력하십시오: ");
    scanf("%f",&distance);
    printf("시간을 입력하십시오: ");
    scanf("%f",&hour);
    printf("분을 입력하십시오: ");
    scanf("%f",&minute);
    printf("초를 입력하십시오");
    scanf("%f",&second);
    v = distance * mtokm / (hour + minute*minutetohour + second*secondtohour) ;
    printf("속도 :%fkm/h",v);
    
    return 0;
}
```

```python
%%shell
gcc HW3-8.c -o output
./output
```
