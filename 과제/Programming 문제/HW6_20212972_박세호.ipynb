# # 6-15

```python
%%writefile 6-15.c

# include <stdio.h>

int main (void) {

    int remain, delt;
    remain = 1000;
    
    printf("초기 연료량: %d\n", remain);

    while(1) {
        printf("연료 충전은 +, 소모는 -로 입력해주세요: ");
        scanf("%d", &delt);

        remain += delt;
        printf("현재 남아있는 양은 %d입니다.\n", remain);

        if (remain <= 100) {
            printf("<경고> 연료가 10%% 이하입니다.");
            break;
        }
    }

    return 0;
}
```

```python
%%shell
gcc 6-15.c -o output
./output
```

# # 6-18

```python
%%writefile 6-18.c

# include <stdio.h>
# include <stdlib.h>
# include <time.h>

int main (void) {

    srand(time(NULL));

    int mon, remain, goal, success, fail;
    int i;

    mon = 50;
    goal = 250;
    success = fail = 0;

    printf("초기 금액 $%d \n", mon);
    printf("목표 금액 $%d \n", goal);

    for(i = 0; i < 1000; i++) {
        remain = mon;

        while(1) {

            if ((double)rand() / RAND_MAX <= 0.5) {
                remain++;
            }

            else {
                remain--;
            }

            if (remain == goal) {
                success++;
                //printf("목표 금액 달성!\n");
                break;
            }

            else if (remain == 0) {
                fail++;
                //printf("돈을 모두 잃었습니다!\n");
                break;
            }
        }
    }

    printf("%d 중의 %d번 승리\n", success+fail, success);
    printf("이긴 확률 = %2.2f \n", ((float)success/(success+fail) * 100));

    return 0;
}
```

```python
%%shell
gcc 6-18.c -o output
./output
```

# # 6-19

```python
%%writefile 6-19.c

# include <stdio.h>
# include <math.h>

int main (void) {
    double sum;
    int n, i;
    sum = 0;

    printf("반복 횟수: ");
    scanf("%d", &n);

    for(i = 0; i < n; i++) {
        sum += 4 / ( 2.0*i + 1 ) * pow(-1, i);
    }

    printf("Pi = %lf",sum);

    return 0;
}
```

```python
%%shell
gcc 6-19.c -o output -lm
./output
```
