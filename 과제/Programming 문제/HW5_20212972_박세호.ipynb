# #5-7

```python
%%writefile 5-7.c

# include <stdio.h>

int main (void) {
    int x, y, z, max, sum;
    printf("삼각형의 세변을 입력하시오: ");
    scanf("%d %d %d",&x, &y, &z);

    max = (x < y) ? y : x;
    max = (max < z) ? z : max;

    sum = x + y + z;

    if (max >= sum - max) {
        printf("유효한 삼각형이 아닙니다. \n");
    }

    else {
        if (x == y && y == z) {
            printf("정삼각형");
        }
        else if (x == y || y == z || z == x) {
            printf("이등변삼각형");
        }
        else {
            printf("그냥 삼각형");
        }
    }   
    

    return 0;
}
```

```python
%%shell
gcc 5-7.c -o output
./output
```

# #5-8

```python
%%writefile 5-8.c

# include <stdio.h>

int main (void) {
    float x, y;
    float a = 0.08, b = 0.17, c = 0.26, d = 0.35 ;

    printf("과세 표준을 입력하십시오(만원):");
    scanf("%f",&x);

    if (x <= 1000) {
        y = x * a;
    }
    else if (x <= 4000) {
        y = 1000 * a + (x - 1000) * b;
    }
    else if (x <= 8000) {
        y = 1000 * a + 3000 * b + (x - 4000) * c;
    }
    else {
        y = 1000 * a + 3000 * b + 4000 * c + (x - 8000) * d;
    }
    
    printf("소득세는 %.0f만원입니다.",y);
    
    return 0;
}
```

```python
%%shell
gcc 5-8.c -o output
./output
```

# #5-9

```python
%%writefile 5-9.c

# include <stdio.h>

int main (void) {
    int x,y;
    char op;

    printf("수식을 입력하십시오: ");
    scanf("%d %c %d",&x, &op, &y);

    switch (op) {
        case '+':
            printf("%d\n",x + y);
            break;
        case '-':
            printf("%d\n",x - y);
            break;
        case '*':
            printf("%d\n",x * y);
            break;
        case '/':
            printf("%d\n",x / y);
            break;
        default:
            printf("지원되지 않는 연산자입니다.\n");
    }
    return 0;
}
```

```python
%%shell
gcc 5-9.c -o output
./output
```

```python
%%writefile 5-10.c

# include <stdio.h>

int main (void) {
    char grade;

    printf("성적을 입력하십시오: ");
    scanf("%c", &grade);

    switch (grade) {
        case 'A':
            printf("아주 잘했어요!");
            break;
        case 'B':
            printf("좋습니다.");
            break;
        case 'C':
            printf("만족스럽습니다.");
            break;
        case 'D':
            printf("더 노력해보세요.");
            break;
        case 'F':
            printf("안타깝습니다.");
            break;
        default:
            printf("유효한 성적을 입력하세요.");
    }
    return 0;
}
```

```python
%%shell
gcc 5-10.c -o output
./output
```
